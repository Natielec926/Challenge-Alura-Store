### Importação dos dados



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
#FATURAMENTO TOTAL DAS LOJAS
print(f"{'-'*5} FATURAMENTO {'-'*5}")
faturamento = {
    "Loja 01": loja["Preço"].sum(),
    "Loja 02": loja2["Preço"].sum(),
    "Loja 03": loja3["Preço"].sum(),
    "Loja 04": loja4["Preço"].sum()
}

for nome_lojas, valor in faturamento.items():
  print(f"{nome_lojas}: R${valor:,.2f}".replace(",", "X").replace(".", ",").replace("X", "."))

#MEDIA DE FATURAMENTO DAS LOJAS
media_faturamento = sum(faturamento.values()) / len(faturamento)
print(f"\nMédia de faturamento das lojas é de R${media_faturamento:,.2f}".replace(",", "X").replace(".", ",").replace("X", "."))


#TOTAL DE VENDAS PARA CADA LOJA
print(f"{'-'*5} CATEGORIA DE PRODUTO {'-'*5}")
total_vendas = {
    "Loja 01": int(loja["Produto"].count()),
    "Loja 02": int(loja2["Produto"].count()),
    "Loja 03": int(loja3["Produto"].count()),
    "Loja 04": int(loja4["Produto"].count())
}

#TOTAL DE VENDAS POR CATEGORIA DE CADA LOJA E EM PORCENTAGEM
vendas_de_categorias = {
    "Loja 01 ": loja["Categoria do Produto"].value_counts(),
    "\nLoja 02 ": loja2["Categoria do Produto"].value_counts(),
    "\nLoja 03 ": loja3["Categoria do Produto"].value_counts(),
    "\nLoja 04 ": loja4["Categoria do Produto"].value_counts()
}

for categoria, quantidade in vendas_de_categorias.items():
  porcentagem = (quantidade / sum(quantidade)) * 100
  print("Vendas por categoria (%)")
  print(f"{categoria} {porcentagem.round(2)}\n")


#VALOR TOTAL DE FRETE POR LOJA
print(f"{'-'*5} FRETE {'-'*5}")
frete = {
    "Loja 01": loja["Frete"].sum(),
    "Loja 02": loja2["Frete"].sum(),
    "Loja 03": loja3["Frete"].sum(),
    "Loja 04": loja4["Frete"].sum()
}

print("\nFRETE\n")
for nome_loja, valor_frete in frete.items():

  print(f"Valor total do frete {nome_loja}:")
  print(f"R${valor_frete:,.2f}\n".replace(",", "X").replace(".", ",").replace("X", "."))

#MEDIA DE FRETE POR LOJA
media_frete = {
    "Loja 01": loja["Frete"].mean(),
    "Loja 02": loja2["Frete"].mean(),
    "Loja 03": loja3["Frete"].mean(),
    "Loja 04": loja4["Frete"].mean()
}

for nome_loja, mediafrete in media_frete.items():
  print(f"O valor médio do frete da {nome_loja}")
  print(f"R${mediafrete:.2f}\n".replace(",", "X").replace(".", ",").replace("X", "."))


#VERIFICAR QUANTIDADE DE VENDAS REALIZADAS POR CADA TIPO DE PAGAMENTO
print(f"{'-'*5} TIPO DE PAGAMENTO {'-'*5}")
tipo_de_pagamento = {
    "Loja 01": loja["Tipo de pagamento"].value_counts(),
    "Loja 02": loja2["Tipo de pagamento"].value_counts(),
    "Loja 03": loja3["Tipo de pagamento"].value_counts(),
    "Loja 04": loja4["Tipo de pagamento"].value_counts()
}

for nome_loja, tipopagamento in tipo_de_pagamento.items():
  print(f"{nome_loja} \n{tipopagamento}\n")



#MEDIA DE AVALIÇÃO DAS LOJAS
print(f"{'-'*5} AVALIAÇÃO {'-'*5}")
media_avaliação_lojas = {
    "Loja 01:": loja["Avaliação da compra"].mean(),
    "Loja 02:": loja2["Avaliação da compra"].mean(),
    "Loja 03:": loja3["Avaliação da compra"].mean(),
    "Loja 04:": loja4["Avaliação da compra"].mean(),

}

for avaliação, media_nota in media_avaliação_lojas.items():
  print(f"{avaliação} {media_nota.round(3)}")
```

```python
# Gráfico do faturamento
plt.figure(figsize=(8, 5))
cores = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728"]
plt.bar(faturamento.keys(), faturamento.values(), color=cores, linewidth=0.5)
plt.title("Faturamento total por loja")
plt.xlabel("Loja")
plt.ylabel("Faturamento (R$)")


for i, (nome_loja, valor) in enumerate(faturamento.items()):
    plt.text(i, valor + 100, f"R${valor:,.2f}".replace(",", "X").replace(".", ",").replace("X", "."),
             ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
#Gráfico por categoria
vendas_de_categorias = {
    "Loja 01": loja["Categoria do Produto"].value_counts(),
    "Loja 02": loja2["Categoria do Produto"].value_counts(),
    "Loja 03": loja3["Categoria do Produto"].value_counts(),
    "Loja 04": loja4["Categoria do Produto"].value_counts()
}

df = pd.DataFrame(vendas_de_categorias).fillna(0)
df_percentual = df.apply(lambda col: (col / col.sum()) * 100)

fig, ax = plt.subplots(figsize=(22, 8))
df_percentual.plot(kind='bar', ax=ax)

plt.title("Distribuição de Vendas por Categoria (%)")
plt.ylabel("Porcentagem de Vendas (%)")
plt.xlabel("Categoria do Produto")
plt.xticks(rotation=45)
plt.legend(title="Lojas")

for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.annotate(f'{height:.1f}%',
                    (p.get_x() + p.get_width() / 2, height),
                    ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```

```python
#Gráfico de custo total de frete por loja
lojas = list(frete.keys())
valores_frete = list(frete.values())

plt.figure(figsize=(8, 6))
cores = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728"]
plt.bar(lojas, valores_frete, color= cores, linewidth=0.5)

for i, valor in enumerate(valores_frete):
    plt.text(i, valor + (valor * 0.01), f"R${valor:,.2f}".replace(",", "X").replace(".", ",").replace("X", "."),
             ha='center', fontsize=10, color='black')

plt.title("Custo Total de Frete por Loja", fontsize=14)
plt.xlabel("Lojas", fontsize=12)
plt.ylabel("Valor do Frete (R$)", fontsize=12)

plt.gca().get_yaxis().set_major_formatter(
    plt.matplotlib.ticker.FuncFormatter(lambda x, _: f"R${x:,.2f}".replace(",", "X").replace(".", ",").replace("X", "."))
)

plt.tight_layout()
plt.show()
```

```python
#Gráfico de frete médio
plt.subplot(1, 2, 2)
cores = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728"]
plt.bar(media_frete.keys(), media_frete.values(), color=cores, linewidth=0.5)
for i, valor in enumerate(media_frete.values()):
    plt.text(i, valor + (valor * 0.01), f"R${valor:,.2f}".replace(",", "X").replace(".", ",").replace("X", "."),
             ha='center', fontsize=10)
plt.title("Frete Médio por Loja", fontsize=14)
plt.xlabel("Lojas", fontsize=12)
plt.ylabel("Valor Médio do Frete (R$)", fontsize=12)

plt.tight_layout()
plt.show()
```

```python
#Gráfico de tipo de pagamentos
df = pd.DataFrame(tipo_de_pagamento).fillna(0)
df_percentual = df.apply(lambda col: (col / col.sum()) * 100)

fig, ax = plt.subplots(figsize=(22, 8))
df_percentual.plot(kind='bar', ax=ax)

plt.title("Distribuição de Tipos de Pagamento (%)")
plt.ylabel("Porcentagem de Pagamentos (%)")
plt.xlabel("Tipo de Pagamento")
plt.xticks(rotation=45)
plt.legend(title="Lojas")

for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.annotate(f'{height:.1f}%',
                    (p.get_x() + p.get_width() / 2, height),
                    ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```

```python
# Gráfico de avaliação por loja
fig, ax = plt.subplots(figsize=(8, 5))
cores = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728"]
ax.bar(media_avaliação_lojas.keys(), media_avaliação_lojas.values(), color=cores, linewidth=0.5)

ax.set_title("Média de Avaliação por Loja")
ax.set_xlabel("Loja")
ax.set_ylabel("Média da Avaliação")
ax.set_ylim(0, 5)  # Ajustando o limite do eixo para escala de avaliação padrão (0-5)


# Adicionando valores acima das barras
for i, (loja, valor) in enumerate(media_avaliação_lojas.items()):
    ax.text(i, valor + 0.1, f"{valor:.2f}", ha='center', va='bottom', fontsize=9)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
